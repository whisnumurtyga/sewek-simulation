```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter
m = 5  # Massa benda dalam kg
g = 9.8  # Percepatan gravitasi dalam m/s^2
koefisien_gesek = 0.2  # Koefisien gesek antara tali dan kotak
panjang_tali = 5  # Panjang tali dalam meter

# Menghitung gaya gesek pada tali
Fgesek = m * g * koefisien_gesek

# Gaya horizontal sebelum mempertimbangkan gaya gesek
Fh_sebelum = m * g

# Memperbarui gaya horizontal dengan mempertimbangkan gaya gesek
if Fh_sebelum >= Fgesek:
    Ftambahan = Fgesek
else:
    Ftambahan = Fh_sebelum

# Gaya horizontal setelah memperbarui nilai Ftambahan
Fh_setelah = Fh_sebelum - Ftambahan

# Visualisasi
labels = ['Gaya Horizontal Sebelum', 'Gaya Gesek', 'Gaya Horizontal Setelah', 'Perasaan Berat Tambahan']
forces = [Fh_sebelum, Fgesek, Fh_setelah, Ftambahan]

plt.bar(labels, forces)
plt.xlabel('Komponen Gaya')
plt.ylabel('Nilai Gaya (Newton)')
plt.title('Perhitungan Gaya Gesek pada Tali dan Gaya Horizontal Setelah Memperhatikan Gaya Gesek')
plt.show()
```
