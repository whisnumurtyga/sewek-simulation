```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def perform_attack(attack_prob, crit_prob):
    # Generate random number to determine attack type
    attack_type = random.random()

    if attack_type < attack_prob:
        return "Normal Attack"
    elif attack_type < attack_prob + crit_prob:
        return "Critical Attack"
    else:
        return "Missed Attack"
```

```python
# Define the probabilities and health for character 1 and character 2
health_char1 = 100
attack_prob_char1 = 0.7   # Character 1 has a 70% chance of a normal attack
crit_prob_char1 = 0.6     # Character 1 has a 30% chance of a critical attack

health_char2 = 100
attack_prob_char2 = 0.6   # Character 2 has a 60% chance of a normal attack
crit_prob_char2 = 0.7     # Character 2 has a 40% chance of a critical attack

# Set the number of rounds for the duel simulation
num_rounds = 10

# Set the number of simulations
num_simulations = 100
```

```python

# Lists to store attack counts and damage dealt by each character
attack_counts_char1 = np.zeros(num_rounds + 1)
damage_dealt_char1 = np.zeros(num_rounds + 1)
attack_counts_char2 = np.zeros(num_rounds + 1)
damage_dealt_char2 = np.zeros(num_rounds + 1)
```

```python
# Create the stacked bar chart for attacks and damage dealt for both characters
plt.figure(figsize=(12, 8))
width = 0.35
x = np.arange(num_rounds + 1)

p1 = plt.bar(x - width/2, attack_counts_char1, width, label='Character 1 Attacks')
p2 = plt.bar(x + width/2, attack_counts_char2, width, label='Character 2 Attacks')
p3 = plt.bar(x - width/2, damage_dealt_char1, width, bottom=attack_counts_char1, label='Damage Dealt (Character 1)')
p4 = plt.bar(x + width/2, damage_dealt_char2, width, bottom=attack_counts_char2, label='Damage Dealt (Character 2)')

plt.xlabel('Rounds')
plt.ylabel('Count')
plt.title('Number of Attacks and Damage Dealt for Both Characters')
plt.legend()
plt.grid(True)
plt.show()
```
