```python
import pandas as pd
import random

# Buat daftar nilai untuk kolom 'from'
from_values = ['unesa', 'itts', 'other']

# Generate fake data
fake_data = []
customer_status = {from_value: 'out' for from_value in from_values}  # Catatan status pelanggan setiap grup 'from'

for hour in range(8, 21):  # Waktu dari jam 8 pagi hingga 8 malam (24 jam)
    for minute in range(0, 60, 15):  # Rentang 15 menit
        time = f"{hour:02d}:{minute:02d}"
        
        # Pilih status secara acak, tapi pastikan "out" hanya muncul jika sebelumnya adalah "in"
        if any(status == 'in' for status in customer_status.values()):
            status = random.choice(['in', 'out'])
        else:
            status = 'in'
            
        # Jika status "out" muncul, pilih secara acak dari grup 'from' yang memiliki status "in"
        if status == 'out':
            from_value_options = [from_value for from_value, status in customer_status.items() if status == 'in']
            from_value = random.choice(from_value_options)
        else:
            from_value = random.choice(from_values)  # Nilai acak untuk kolom 'from'
        
        # Update status pelanggan grup 'from' berdasarkan status saat ini
        customer_status[from_value] = status
        
        # Jumlah pelanggan saat ini pada setiap grup 'from' dengan status "in"
        num = list(customer_status.values()).count('in')
        
        # Tambahkan data palsu ke dalam daftar fake_data
        fake_data.append({'time': time, 'num': num, 'status': status, 'from': from_value})

# Konversi data palsu ke dalam pandas DataFrame
df = pd.DataFrame(fake_data)

# Tampilkan data dalam DataFrame
print(df)
```
