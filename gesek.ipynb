```python
import numpy as np
import matplotlib.pyplot as plt

# Konstanta dan parameter
g = 9.8         # Percepatan gravitasi (m/s^2)
c = 0.1         # Koefisien gesek kinetis
m = 1.0         # Massa benda (kg)
v0 = 10.0       # Kecepatan awal benda (m/s)
t_total = 100.0  # Total waktu simulasi (s)
dt = 1 # Step waktu (s)

# Array untuk menyimpan hasil perhitungan
time_array = np.arange(0, t_total, dt)
position_array = np.zeros(len(time_array))
velocity_array = np.zeros(len(time_array))
acceleration_array = np.zeros(len(time_array))
force_gravity_array = np.zeros(len(time_array))
force_friction_array = np.zeros(len(time_array))
total_distance = 0.0

# Menghitung pergerakan benda menggunakan metode numerik (Euler)
for i in range(len(time_array)):
    # Menghitung gaya gravitasi
    force_gravity = m * g

    # Menghitung gaya gesek kinetis
    force_friction = c * m * g

    # Menghitung gaya total yang dialami benda
    force_total = force_gravity - force_friction

    # Menghitung percepatan benda
    acceleration = force_total / m

    # Menghitung kecepatan benda
    if i == 0:
        velocity = v0
    else:
        velocity = velocity_array[i - 1] + acceleration * dt

    # Menghitung posisi benda
    position = 0.5 * acceleration * time_array[i]**2 + v0 * time_array[i]

    # Menyimpan hasil perhitungan pada array
    position_array[i] = position
    velocity_array[i] = velocity
    acceleration_array[i] = acceleration
    force_gravity_array[i] = force_gravity
    force_friction_array[i] = force_friction

    # Menghentikan simulasi jika benda berhenti
    if velocity <= 0:
        total_distance = position_array[i]
        break

# Plot grafik pergerakan benda
plt.figure(figsize=(10, 6))
plt.plot(time_array, position_array, label='Posisi Benda')
plt.plot(time_array, velocity_array, label='Kecepatan Benda')
plt.plot(time_array, acceleration_array, label='Percepatan Benda')
plt.xlabel('Waktu (s)')
plt.ylabel('Nilai')
plt.title('Pergerakan Benda')
plt.legend()
plt.grid(True)
plt.show()

# Tampilkan hasil simulasi
print("Massa Benda:", m, "kg")
print("Kecepatan Awal Benda:", v0, "m/s")
print("Total Jarak Tempuh Benda:", total_distance, "m")
```
