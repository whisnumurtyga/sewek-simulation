```python
import random

# Inisialisasi variabel
max_capacity = 50  # Kapasitas maksimal tempat makan
density_threshold = 0.8  # Ambang batas kepadatan
current_visitors = 0  # Jumlah pengunjung saat ini

# Pola kedatangan pengunjung
hourly_arrivals = [10, 15, 20, 25, 30, 35, 40, 45, 40, 35, 30, 25, 20, 15, 10, 5]  # Jumlah pengunjung per jam

# Inisialisasi list untuk menyimpan data simulasi
density_data = []  # Data kepadatan pengunjung

# Simulasi untuk jangka waktu tertentu
sim_duration = 60  # Durasi simulasi dalam menit

for minute in range(sim_duration):
    # Hitung jumlah pengunjung baru pada setiap menit berdasarkan pola kedatangan
    current_hour = minute // 60  # Jam saat ini
    new_visitors = hourly_arrivals[current_hour]

    # Generate jumlah pengunjung baru dengan variasi acak
    new_visitors += random.randint(-2, 2)

    # Batasi jumlah pengunjung baru dengan kapasitas tempat makan
    if current_visitors + new_visitors > max_capacity:
        new_visitors = max_capacity - current_visitors if current_visitors < max_capacity else 0

    current_visitors += new_visitors

    # Hitung kepadatan pengunjung
    density = current_visitors / max_capacity
    density_data.append(density)

    # Cek apakah kepadatan mencapai ambang batas
    if density >= density_threshold:
        print(f"Kepadatan tinggi! {new_visitors} pengunjung bungkus makanan.")
        # Asumsikan semua pengunjung baru memesan makanan untuk dibungkus
        current_visitors -= new_visitors

    # Tampilkan informasi simulasi pada setiap menit
    print(f"Menit {minute+1}: Pengunjung baru: {new_visitors}, Total pengunjung: {current_visitors}, Kepadatan: {density}")

# Visualisasi data kepadatan pengunjung
import matplotlib.pyplot as plt

plt.plot(range(sim_duration), density_data)
plt.axhline(y=density_threshold, color='r', linestyle='--')
plt.xlabel('Menit')
plt.ylabel('Kepadatan Pengunjung')
plt.title('Simulasi Kepadatan Pengunjung di Tempat Makan')
plt.show()
```
